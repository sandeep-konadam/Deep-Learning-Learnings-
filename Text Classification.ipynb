```python
import numpy as np
import pandas as pd
df=pd.read_csv('Book4.csv',encoding='ISO-8859-1')
df.head
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report
```

```python
documents = df["Text"]
labels=df["Sentiment"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(documents, labels, test_size=0.2, random_state=42)
```

```python
tfidf_vectorizer = TfidfVectorizer(stop_words='english', max_features=5000)
```

```python
X_train_tfidf = tfidf_vectorizer.fit_transform(X_train)
```

```python
X_test_tfidf = tfidf_vectorizer.transform(X_test)
```

```python
classifier = MultinomialNB()
```

```python
classifier.fit(X_train_tfidf, y_train)
```

```python
predictions = classifier.predict(X_test_tfidf)
```

```python
accuracy = accuracy_score(y_test, predictions)
report = classification_report(y_test, predictions)

print(f"Accuracy: {accuracy}")
print("Classification Report:")
print(report)
```

```python
# Sample sentences for testing
test_sentences = ["I love this product!", "bad", "Amazing experience so far."]

# Transform the test sentences using the same TfidfVectorizer
test_sentences_tfidf = tfidf_vectorizer.transform(test_sentences)

# Make predictions on the test sentences
predictions_test = classifier.predict(test_sentences_tfidf)

# Display the predictions
for sentence, prediction in zip(test_sentences, predictions_test):
    sentiment = "Positive" if prediction == 1 else "Negative"
    print(f"Sentence: {sentence}\nPredicted Sentiment: {sentiment}\n")
```
