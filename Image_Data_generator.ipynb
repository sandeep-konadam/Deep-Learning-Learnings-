```python
import tensorflow
from tensorflow import keras
from keras.preprocessing.image import ImageDataGenerator,array_to_img,img_to_array,load_img
```

```python
datagen=ImageDataGenerator(
    samplewise_center=True,
    horizontal_flip=True,
    vertical_flip=True,
    height_shift_range=0.2,
    rotation_range=20,
    shear_range=0.3,
    zoom_range=0.4,
    channel_shift_range=0.5
)
```

```python
img=load_img('photo-1614027164847-1b28cfe1df60.jpg')
x=img_to_array(img)
x=x.reshape((1,)+x.shape)
```

```python
i=0
for batch in datagen.flow(x,batch_size=1,save_to_dir="Preveiw",save_format='jpeg',save_prefix='Lion'):
    i+=1
    if i>20:
      break
```
